```python
import pandas as pd
import matplotlib.pyplot as plt

tiger_population = {'Andhra Pradesh' : 62, 'Arunanchal Pradesh' : 29, 'Assam' : 182, 'Bihar' : 31, 'Chhatisgarh' : 19, 'Goa' : 2, 'jharkhand' : 3, 'Karnatak' : 524, 'Keral' : 183, 'Madhya Pradesh' : 785, 'Maharastra' : 444, 'Orisha' : 20, 'Rajasthan' : 88, 'Tamil Nadu' : 306, 'Telangana' : 21, 'Uttar Pradesh' : 205, 'Uttrakhand' : 560, 'Bengal' : 131}

tiger_dataframe = pd.DataFrame(list(tiger_population.items()), columns=['State','population'])
tiger_dataframe.sort_values(by='population', ascending=False, inplace=True)

plt.figure(figsize=(10,8), dpi=300)
bars = plt.barh(tiger_dataframe['State'], tiger_dataframe['population'], color='skyblue')

for bar in bars:
    plt.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/3, 
             str(int(bar.get_width())), 
            )

plt.xlabel('States', fontsize=12)
plt.ylabel('Tiger Population', fontsize=12)
plt.title(' Tiger Population in India', fontsize=14)

plt.savefig('tiger_populatioin.png', dpi=300, bbox_inches='tight')
plt.show()
```
