## Modeling

```python
# split data into trainging and testing data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=2)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# above warning is i cannot use gpu for the tensorflow library as tensorflow use cuda and cuda does not support intel gpu
```

```python
tf.random.set_seed(3)
```

```python
# create model

# Seqential is a Class of tensorflow which help in creating neural network. TensorFlow use high level abstract inplementation of neural network.
# input layer, Flatten function takes columns(properties) value and convvert them into 1D array
# dense layer, this layers requires no of neuron in this layer and activation function
# output layer, this layer provide final output of the training set. The no of set depend on the no of classes the target value have.
# In our case we have two value 0 and 1. For multiple classes we use softmax function

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

# Compiling the neural Network
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid'),
])

model2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# training
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
history2 = model2.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
# Standarize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
history3 = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
history4 = model2.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```
